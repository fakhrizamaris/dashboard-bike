## Exploratory Data Analysis (EDA)

### Explore ...

```python
hour_df.describe()
```

```python
day_df.describe()
```

```python
all_df = pd.merge(
    left=hour_df,
    right=day_df,
    how="right",
    left_on="instant",
    right_on="instant"
)
all_df.head()
```

```python
all_df.describe()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Set plot style
sns.set(style="whitegrid")


plt.figure(figsize=(10, 6))
sns.boxplot(x='weathersit_y', y='cnt_y', data=all_df)
plt.title('Dampak Kondisi Cuaca pada jumlah penyewaan sepeda')

#Dari Visualisai yang didapat saya membuat kategori cuaca menjadi Cerah, Berkabut, dan Hujan agar Visualisasi lebih mudah dipahami
plt.xlabel('Kondisi Cuaca (1.Cerah 2.Berkabut 3.Hujan)')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.show()
```

### Pertanyaan 2:

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))

all_df[['casual_x', 'registered_x']].sum().plot(kind='bar', color=['purple', 'green'])

plt.title('Perbandingan Penyewaan Sepeda: Casual vs Registered', fontsize=16)
plt.xlabel('Tipe Pengguna', fontsize=12)
plt.ylabel('Jumlah Penyewaan', fontsize=12)
plt.show()
```

## Conclusion

- Pengaruh Cuaca mempengaruhi jumlah penyewaan sepeda, itu terlihat ketika saat kondisi cuaca dengan kategori 1(Cerah) jumlah penyewaan sepeda berkisar 3700 hingga 6400 lebih tinggi dibandingkan saat kondisi cuaca dengan kategori 2 dan 3.
- Perbandingan jumlah penyewa casual dan jumlah penyewa terdaftar terlihat signifikan dengan perbedaan yang jauh lebih banyak penyewa terdaftar. Hal ini dapat disimpulkan bahwa penyewaan sepeda sukses untuk mempromosikan pelanggan agar mendaftar pada bisnisnya.
